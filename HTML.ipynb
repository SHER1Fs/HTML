# **Form :**


* < form > consist of :
 
   * **< input > :** inline self-closing tag 
  
     * input attributes : 
 
         * a ) type : text , password , submit , email , hidden , reset , color , range(value,min,max and step) , number .

           * radio : Creates a radio button for mutually exclusive choices within a group **(only one choise)**
           * checkbox : Creates a checkbox for binary choices (checked or unchecked) **(multiple choises)**

           * file :  Creates a file input field for file uploads .

           * search :  Creates a search input field

           * URL : Creates an input field for entering a URL only (do not accept regular text)

           * date : Creates an input field for entering a date.
         
         * b ) required : the input must be filled

         * c ) placeholder : describe the required input same as the label tag , but written inside the input box . 

         * d ) value : default value will be displayed before entering any data 

         * e ) readonly :  it makes the input control read-only. The user can see the value but cannot modify it. 

         * f ) disabled :it disables the input control, making it non-editable and preventing the user from interacting with it.

         * g ) minlength : Specifies the minimum number of characters required in a text input 

         * h ) id : to link the input with the label (must be unique) 

         
     * **< label > :** describe the required input

       * label attribute : 

         * for="id" : to link it with the input 

     * **< select > :** used to create a dropdown list , consists of many options Each option is defined by an **< option >** element nested inside the < select > element , those many option can be grouped using **< optgroup >** element

       * < option > attributes : 

         * value :  is what gets sent to the server when the form is submitted, while the text content between the opening and closing < option > tags is what is displayed to the user in the dropdown list.  

       * < optgroup > attribute :

         * label : the group name  
      
       * select attributes :

         * name : used to identify the field when the form is submitted
         * id : used to uniquely identify the dropdown list 
         * multiple : to choose more than one option 

       * < list > element can do  all < select > functions and can search in it not only choosing            


     * **< textarea > :** Creates a multi-line text input field    

   * form attributes : 

     * 1 ) action : Specifies the URL to which the form data should be sent for processing  

     * 2 ) name : Provides a name for the form, which can be used to identify the form in scripting .

     * 3 ) method : pecifies the HTTP method to be used when sending form data. The two most common methods are "GET" and "POST" . **post** hide the sent data    

     * 4 ) novalidate :disables browser validation of form elements. 

     * 5 ) target : Specifies where to display the response received after submitting the form. (_top - _blank - _self - _parent)

     * 6 ) enctype : Specifies how the form data should be encoded before sending it to the server

       * a ) application/x-www-form-urlencoded (default) : It URL-encodes the form data and appends it to the form's action URL as query parameters. Spaces are replaced with the "+" character, and special characters are URL-encoded.

       * b ) multipart/form-data : used when you are uploading files through the form. It is more efficient for binary data like images or documents.

       * c ) text/plain : sends the form data as plain text with no special formatting

# **Event Handlers :**

 * attributes that define the behavior of elements in response to various events

 * it is typically a JavaScript code that will be executed when the specified event occurs.

 * EX : 

 
    * **1 ) onclick :**

         * Triggered when the user clicks on an element. 

    * **2 ) onkeyup :**    

         * Triggered when a key is released 

    * **3 ) onkeydown :**

         * Triggered when a key is pressed

    * **4 ) onkeypress :**

         * Triggered when a key is pressed or released 

    * **5 ) onmouseover :**

         * Triggered when the mouse pointer enters an element.        

    * **6 ) onmouseout :**

         * Triggered when the mouse pointer leaves an element. 

    * **7 ) onerror :**

         * Triggered when an erro occurs .

    * **8 ) onload :**

         * Triggered when the browser finishes loading the document .  

    * **9 ) onfocus :**

         * Triggered when an element gains focus.
